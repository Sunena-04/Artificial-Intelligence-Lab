```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Creating the dataset
data = {
    'Fruit': ['Apple', 'Lemon', 'Cherry', 'Avocado', 'Berry', 'Olive', 'Banana', 'Grapefruit', 
              'Tomato', 'Lime', 'Blueberry', 'Blackberry', 'Poisonberry'],
    'Color': ['Red', 'Yellow', 'Red', 'Green', 'Purple', 'Green', 'Yellow', 'Pink', 
              'Red', 'Green', 'Blue', 'Black', 'Purple'],
    'Weight': [150, 120, 10, 200, 5, 5, 110, 250, 100, 60, 1, 3, 3],
    'Edible': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No']
}
# Convert to DataFrame
df = pd.DataFrame(data)
# We will only use the 'Weight' column for clustering
X = df[['Weight']].values
# Applying KMeans clustering with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
# Getting the centroids
centroids = kmeans.cluster_centers_
# Output the clusters and centroids
print("Cluster Assignments:\n", df[['Fruit', 'Weight', 'Cluster']])
print("\nCentroids of the Clusters:")
print(centroids)
# Plotting the data points and centroids for better visualization
plt.scatter(df['Weight'], np.zeros_like(df['Weight']), c=df['Cluster'], cmap='viridis')
plt.scatter(centroids[:, 0], np.zeros_like(centroids[:, 0]), color='red', marker='x', s=200, label='Centroids')
plt.title('K-Means Clustering (K=2)')
plt.xlabel('Weight')
plt.legend()
plt.show()
```
